```python
import pandas as pd
import ast
from datasets import load_dataset
import matplotlib.pyplot as plt
datasets=load_dataset('lukebarousse/data_jobs')
df=datasets['train'].to_pandas()
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df['job_skills']=df['job_skills'].apply(lambda skill_list:ast.literal_eval(skill_list)
                                        if pd.notna(skill_list) else skill_list
                                        )
```

```python
df[['job_title_short','job_skills']].head()
```

```python
df_exploded=df.explode('job_skills')
```

```python
df_exploded
```

```python
skill_count=df_exploded.groupby(['job_skills','job_title_short']).size()
```

```python
skill_count
```

```python
df_skills_count=skill_count.reset_index(name='skill_count') #changes the series to dataframe
```

```python
df_skills_count=df_skills_count.sort_values(by='skill_count',ascending=False)
df_skills_count
```

```python
job_title='Data Analyst'
top_skills=10
```

```python
df_skills_final=df_skills_count[df_skills_count['job_title_short']==job_title].head(top_skills)
```

```python
df_skills_final
```

```python
df_skills_final.plot(kind='barh',x='job_skills',y='skill_count')
plt.gca().invert_yaxis()
plt.legend().set_visible(False)
```
