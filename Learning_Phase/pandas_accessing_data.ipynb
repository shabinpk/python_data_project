```python
datasets=load_dataset('lukebarousse/data_jobs')
dft=datasets['train'].to_pandas()
```

```python
top_countries=dft['job_country'].value_counts().head(6).index

df_job_country_salary=dft.pivot_table(
    values='salary_year_avg',
    index='job_country',
    columns='job_title_short',
    aggfunc='median'
)
df_job_country_salary=df_job_country_salary.loc[top_countries]
df_job_country_salary
#here used top_countries as index and assigned to job_country)salary
```

```python
job_titles=['Data Analyst','Data Engineer','Data Scientist']
df_job_country_salary=df_job_country_salary[job_titles]
df_job_country_salary
#When you use [] after a variable, you're effectively filtering or selecting specific elements from that variable
```

```python
df_job_country_salary.plot(kind='bar')
plt.ylabel('Median Salary in USD')
plt.xlabel('')
plt.title('Median Salary by country & job titles')
plt.xticks(rotation=45,ha='right')
plt.show()
```
