```python
import pandas as pd
import ast
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt


# Loading Data
datasets=load_dataset('lukebarousse/data_jobs')
df=datasets['train'].to_pandas()


# Data Cleanup
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df['job_skills']=df['job_skills'].apply(lambda skill_list:ast.literal_eval(skill_list) if pd.notna(skill_list) else skill_list)
```

```python
df_da_us=df[(df['job_country']=='United States') & (df['job_title_short']=='Data Analyst')].copy()
df_da_us.head()
```

```python
df_da_us['job_posted_month_no']=df_da_us['job_posted_date'].dt.month
```

```python
df_da_us_exploded=df_da_us.explode('job_skills')
df_da_us_exploded
```

```python
df_da_us_pivot=df_da_us_exploded.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)
df_da_us_pivot.loc['Total']=df_da_us_pivot.sum()
df_da_us_pivot=df_da_us_pivot[df_da_us_pivot.loc['Total'].sort_values(ascending=False).index].drop('Total')
df_da_us_pivot
```

```python
da_totals=df_da_us.groupby('job_posted_month_no').size()

da_totals
```

```python
df_da_us_percent= df_da_us_pivot.div(da_totals, axis=0)*100
df_da_us_percent
```

```python
#change month number to month name
df_da_us_percent=df_da_us_percent.reset_index()
df_da_us_percent['job_posted_month']=df_da_us_percent['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_da_us_percent=df_da_us_percent.set_index('job_posted_month')
df_da_us_percent=df_da_us_percent.drop(columns=['job_posted_month_no'])
```

```python
df_da_us_percent
```

```python
df_plot=df_da_us_percent.iloc[:,:5]
sns.lineplot(data=df_plot,dashes=False,palette='tab10')
sns.set_theme(style='ticks')
plt.title('Top 5 Skills in Data Analyst Job Postings in the US')
plt.ylabel('Likelihood of Job Postings (%)')
plt.xlabel('2023')
sns.despine()
plt.legend().remove()
ax=plt.gca()
from matplotlib.ticker import PercentFormatter
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))
for i in range(5):
    plt.text(11.2,df_plot.iloc[-1,i],df_plot.columns[i])
```
